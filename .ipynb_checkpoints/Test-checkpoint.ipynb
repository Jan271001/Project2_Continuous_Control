```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
scores = [[0,10],[10,20]]
means = [1,1]
episode_scores = []
for record in scores:
    episode_scores.append(np.mean(record))

fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(len(episode_scores)), episode_scores, label = "scores")
plt.plot(np.arange(len(means)), means, label = "means")
plt.axhline(y=30, color='r', linestyle='--', label="target")
plt.ylabel('Score')
plt.xlabel('Episode')
plt.legend()
plt.show()
```

```python
test = []
test2 = []
for i in range(10):
    test.append(i)
    test2.append(9-i)

mean = np.mean(test)
print(mean)

#plot the scores
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(len(test)), test, label="test")
plt.plot(np.arange(len(test2)), test2, label="test2")
plt.axhline(y=mean, color='r', linestyle='--', label="mean")
plt.ylabel('Score')
plt.xlabel('Episode')
plt.legend()
plt.show()
```
