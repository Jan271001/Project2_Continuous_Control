```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
test = []
test2 = []
for i in range(10):
    test.append(i)
    test2.append(9-i)

mean = np.mean(test)
print(mean)

#plot the scores
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(len(test)), test, label="test")
plt.plot(np.arange(len(test2)), test2, label="test2")
plt.axhline(y=mean, color='r', linestyle='--', label="mean")
plt.ylabel('Score')
plt.xlabel('Episode')
plt.legend()
plt.show()
```
